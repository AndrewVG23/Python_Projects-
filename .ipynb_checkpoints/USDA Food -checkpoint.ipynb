```python
wb = load_workbook(filename = 'ABBREV.xlsx')
ws = wb.active
```

```python
data = ws.values
cols = next(data)[1:]
data = list(data)
idx = [r[0] for r in data]
data = (itr.islice(r, 1, None) for r in data)
df = pd.DataFrame(data, index=idx, columns=cols)
```

```python
#need to normalize

Count = df.iloc[0][1:15]
Name = df.columns[1:15]

r_name = np.arange(len(Name))


plt.figure(figsize=(10,5))
plt.bar(r_name, Count, align='center', alpha=.5)
plt.xticks(r_name, Name)
plt.ylabel('Grams')
plt.title('Butter')
plt.show()
```

```python
colormap = plt.cm.rainbow
plt.figure(figsize=(20,20))
plt.title('Correlations of the Features of Food', y=1.05, size=15)
sns.heatmap(df.corr(),linewidths=0.1,vmax=1.0, square=True, 
            cmap=colormap, linecolor='black', annot=False)
plt.show()
```

Let's do a closer analysis of that high protein, low carb section

```python
prtn = df[df['Protein_(g)']>13]
bdybld = prtn[prtn['Carbohydrt_(g)']<10]
```

```python
colormap = plt.cm.rainbow
plt.figure(figsize=(20,20))
plt.title('Correlations of the Features of "Body Building" Food', y=1.05, size=15)
sns.heatmap(bdybld.corr(),linewidths=0.1,vmax=1.0, square=True, 
            cmap=colormap, linecolor='black', annot=False)
plt.show()
```

```python
colormap = plt.cm.rainbow
plt.figure(figsize=(20,20))
plt.title('Unique Features in the Content of "Body Building Food"', y=1.05, size=15)
sns.heatmap(abs(bdybld.corr())-abs(df.corr()),linewidths=0.1,vmax=1.0, square=True, 
            cmap=colormap, linecolor='black', annot=False)
plt.show()
```

Word Cloud of the short descriptions for the "body building" foods

UNDER CONSRUCTION

```python
bdybld_words = nltk.word_tokenize(' '.join(bdybld['Shrt_Desc'].values))

stopwords = [',','&','brld','var','ln,','ln,0','fat,0','grds','sel','brsd','broilers','fryers','or','grd','all','frsh','bnless','rnd','skn','fat','choic','ln','ckd','raw','nz','imp']

bdybld_words_cleaned = [word for word in bdybld_words if word.lower() not in stopwords]

#' '.join(bdybld_words_cleaned)
```

```python
wordcloud = WordCloud().generate(' '.join(bdybld_words_cleaned))
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```

```python
df1 = df[df['Protein_(g)']<7]
low_prtn = df1[df1['Carbohydrt_(g)']<30]
```

```python
low_prtn_words = nltk.word_tokenize(' '.join(low_prtn['Shrt_Desc'].values))

stopwords = ['ckd','bld','wo','drnd','raw','pk','cnd','sol','wo,','str']

low_prtn_words_clean = [word for word in low_prtn_words if word.lower() not in stopwords]
```

```python
wordcloud = WordCloud().generate(' '.join(low_prtn_words_clean))
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
